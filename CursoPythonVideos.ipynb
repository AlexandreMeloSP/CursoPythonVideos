Calculando Hipotenusa - método tradicional

```python
cateto_oposto = float(input('Informe o valor do cateto oposto: '))
cateto_adjacente =  float(input('Informe o valor do cateto adjacente: '))
hipotenusa = (cateto_oposto ** 2 + cateto_adjacente ** 2) ** (1/2)
display (f'A hipotenusa mede: {hipotenusa}')
```

Aula 8 - Ex. 18 - Calculando seno, coseno e tangente - método mesclado

```python
from math import sin, cos, tan, radians, ceil
angulo = float(input('Informe o ângulo a ser calculado:'))
seno = sin(radians(angulo))
cosseno = cos(radians(angulo))
print(f'O seno de ângulo {angulo} é {seno}')
print(f'O coseno de ângulo {angulo} é {cosseno}')
print(f'A tangente de {angulo} é {tan(radians(angulo))}')
```

Aula Add - While 

```python
valor = int(input('Insira um valor para contar: 	'))
cont = 0
varPar = []
varImpar = []
while cont <= valor:
  if cont % 2 == 0:
    varPar.append(cont)
  else:
    cont % 2 == 1
    varImpar.append(cont)
  cont += 1
if valor % 2 == 0:
  print(f'Números acrescentados à variável par \' {varPar} \'') # Contando até 5
if valor % 2 == 1:
  print(f'Números acrescentados à variável impar \' {varImpar} \'') # Contando até 5
```
